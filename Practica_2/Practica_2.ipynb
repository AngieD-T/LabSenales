## Primer Punto

```python
def EscalonUnitario(n0, n1, n2):
 n = np.arange(n1,n2+1)
 x = (n >= n0).astype(int)
 return [x,n]

x, n = EscalonUnitario(4, 1, 6)
```

## Segundo Punto

```python
def Rampa(n0, n1, n2):
 n = np.arange(n1,n2+1)
 x = np.where(n >= n0, n - n0,0)
 return [x,n]

x, n = Rampa(5, 1, 10)
```

## Cuarto Punto

```python
def func():
    r1, n1 = Rampa(0, 0, 5)
    r2, n2 = Rampa(5, 6, 11)
    r3, n3 = Rampa(10, 12, 17)

    x = np.concatenate((r1,r2,r3))
    n = np.concatenate((n1,n2,n3))
    return([x,n])
```

Item a

```python
def derivada(x,n):
    xprima = np.diff(x)/np.diff(n)
    nprima = []
    for i in range(len(xprima)):
      ntemp = (x[i+1]+x[i])/2
      nprima = np.append(nprima,ntemp)
    return nprima, xprima

x = func()[0]
n = func()[1]
nprima, xprima = derivada(n,x)
```

Item b

```python
plt.subplot(2, 1, 1)
plt.stem(n, x)
plt.title('Función x ')
plt.xlabel('n')
plt.ylabel('x[n]')

plt.subplot(2, 1, 2)
plt.stem(nprima,xprima)
plt.title('Derivada de la Función x')
plt.xlabel('n')
plt.ylabel('x[n]')

plt.show()
```

La derivada en un caso continuo toma la resta de infinitos puntos, mientras que en un caso discreto so toma en puntos finitos

# Quinto Punto

```python
# x(n) =  # n+5 si -5 <= n <= -1
          # 5-n si 0 <= n <= 5
          # n-5 si 6 <= n <= 9
          # 5   si 10 <= n <= 13
          # 10  si n > 13

def RampaInversa(n0, n1, n2):
    n = np.arange(n1, n2 + 1)
    x = np.where(n <= n0, n0 - n, 0)
    return [x, n]

def x(n=0):

    x0,n0 = EscalonUnitario(0, -20, -6)
    x1,n1 = Rampa(-5, -5, -1)
    x2,n2 = RampaInversa(5, 0, 5)
    x3,n3 = Rampa(5, 6, 9)
    x4,n4 = EscalonUnitario(0, 10, 13)
    x5,n5 = EscalonUnitario(0, 14, 17)
    x6,n6= EscalonUnitario(0, 18, 20)

    xs = np.concatenate((x1,x2,x3,5*x4,10*x5))
    ns = np.concatenate((n1+n,n2,n3,n4,n5+n))
    xs2 = np.concatenate((0*x0,x1,x2,x3,5*x4,10*x5,0*x6))
    ns2 = np.concatenate((n0,n1,n2,n3,n4,n5,n6))
    a = 0
    if n < -3:
        a = -n-3
        desplazado = np.zeros(len(ns2)+a)
        ns2 = np.append(ns2,ns2[-1]+a)
    else:
        desplazado = np.zeros_like(ns2)

    start = 15 - n
    end = 38 - n
    desplazado[start:end] = xs

    return [xs,ns,xs2,ns2,desplazado]
```
