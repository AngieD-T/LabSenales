## **CONTEXTO**


#### **Parkinson**

Enfermedad de Parkinson es un tipo de trastorno del movimiento. Ocurre cuando las células nerviosas (neuronas) no producen suficiente cantidad de una sustancia química importante en el cerebro conocida como dopamina. [1]

Actualmente, no es posible diagnosticar el Parkinson antes de la aparición de los síntomas. Se estima que, cuando los primeros signos motores se manifiestan, ya se ha perdido entre el 60% y el 80% de las neuronas afectadas. [2] El diagnóstico es realizado por un neurólogo(especialista en enfermedades del sistema nervioso)  a través de la evaluación de los antecedentes médicos, los síntomas y un examen neurológico y físico. Una vez que los síntomas se presentan, se pueden utilizar técnicas de diagnóstico por imagen, como la tomografía computarizada por emisión de fotón único, conocida como "estudio del transportador de dopamina", así como la ecografía transcraneal y la resonancia magnética, para confirmar la sospecha de Parkinson. También se pueden realizar pruebas de laboratorio, como análisis de sangre, para descartar otras posibles causas de los síntomas. Por lo tanto, el proceso de diagnóstico puede tomar tiempo[3].


#### **Huntington**

La enfermedad de Huntington es un trastorno genético en el cual las neuronas en ciertas partes del cerebro se desgastan o se degeneran, esto causa movimientos aleatorios y sueltos que son involuntarios y no se pueden suprimir. [4]

Para el diagnóstico de esta enfermedad, el neurólogo realiza una entrevista exhaustiva para obtener los antecedentes médicos y familiares del paciente y descartar otras afecciones. Se llevan a cabo exámenes neurológicos, genéticos y físicos que evalúan reflejos, equilibrio, movimiento, tono muscular, audición, marcha y estado mental. Si aún hay dudas sobre el diagnóstico, el médico puede recomendar imágenes cerebrales, como una tomografía computarizada (TC) o, más comúnmente, una resonancia magnética (RM). A medida que la enfermedad avanza, estas exploraciones suelen mostrar una contracción de ciertas áreas del cerebro y un aumento en el tamaño de las cavidades llenas de líquido, llamadas ventrículos. Sin embargo, estos cambios no son definitivos para el diagnóstico, ya que también pueden presentarse en otros trastornos. Es posible que una persona con síntomas tempranos de Huntington tenga resultados normales en una tomografía computarizada o una resonancia magnética. [5]


#### **Esclerosis Lateral Amiotrófica (ELA)**

Es una enfermedad del sistema nervioso que afecta las neuronas del cerebro y la médula espinal, esta hace perder el control muscular. [6]

No existe una sola prueba que pueda diagnosticar con certeza la ELA. El médico lleva a cabo un examen físico y revisa la historia clínica completa del paciente. Un examen neurológico evalúa los reflejos, la fuerza muscular y otras respuestas. Estas pruebas deben realizarse de manera regular para monitorear si los síntomas empeoran con el tiempo. También se pueden realizar pruebas musculares y estudios de diagnóstico por imágenes para descartar otras enfermedades, lo que puede ayudar a respaldar un diagnóstico de ELA.

Dado que no hay un método claro y determinante, el proceso de diagnóstico suele prolongarse, lo que representa un desafío significativo. Es crucial obtener un diagnóstico preciso de ELA lo antes posible, ya que los tratamientos tienden a ser más eficaces en las etapas iniciales de la enfermedad. [7]

#### **Entropía de permutación**

La entropía de permutación es una medida utilizada para cuantificar la complejidad o la irregularidad en series temporales. Se basa en la idea de que, al considerar todas las posibles permutaciones de los valores de una serie, se puede obtener información sobre su estructura y patrones subyacentes. En bioingeniería permite analizar y comprender la complejidad de señales biológicas y fisiológicas, evaluación de tratamientos y detección de anomalías, y la obtención de información valiosa para la investigación y la práctica clínica.

En el caso puntual de análisis de marcha para detección de enfermedades, la entropía de permutación puede ayudar a cuantificar patrones de movimiento menos coordinados y más predecibles, permitiendo distinguir entre patrones de marcha normales y patológicos. También, una disminución en la entropía de permutación puede indicar una menor variabilidad en la marcha, lo que podría estar relacionado con la progresión de la enfermedad.

**Función para cálculo de coeficiente de variación (CV) y desviación estandar (SD)**

```python
def calcular_cv_sd(signal):

  lista_cv = []
  lista_sd = []

  for i in range(64):
    media = np.mean(signal[i][0])
    sd = np.std(signal[i][0])

    lista_sd.append(sd)

    if media != 0:
        cv = (sd / media) * 100
    else:
        cv = np.nan

    lista_cv.append(cv)

  return lista_cv, lista_sd
```

```python
cv_LSTI, sd_LSTI = calcular_cv_sd(filtered_array_LSTI)
cv_RSTI, sd_RSTI = calcular_cv_sd(filtered_array_RSTI)
cv_LSWI, sd_LSWI = calcular_cv_sd(filtered_array_LSWI)
cv_RSWI, sd_RSWI = calcular_cv_sd(filtered_array_RSWI)
```

**Cálculo de entropia de la permutación**

```python
def shannon_entropy(x):

  p , _ = np.histogram(x,bins='fd') #'fd' (estimador de Freedman Diaconis)
  p = p/len(p) #proportion

  h = -p * np.log2(p)

  h[np.isnan(h)] = 0.0
  H = np.sum(h)
  return H
```

```python
def calcular_entr(signal):
    entropia = []

    for i in range(64):
      ent = shannon_entropy(signal[i][0])
      entropia.append(ent)
    return entropia
```

```python
entropy_array_LSTI = calcular_entr(filtered_array_LSTI)
entropy_array_RSTI = calcular_entr(filtered_array_RSTI)
entropy_array_LSWI = calcular_entr(filtered_array_LSWI)
entropy_array_RSWI = calcular_entr(filtered_array_RSWI)
```

##**PRUEBA**



Se analiza únicamente la desviación estandar de las 4 enfermedades, ya que al examinar las gráficas de cada métrica, se llegó a la conclusión que la DS ofrece una mayor discriminación en los datos, arrojando más información valiosa para el estudio.


**1. Prueba de normalidad**

**Ho:** Los datos siguen una distribución normal

**H1:** Los datos  No siguen una distribución normal

```python
df_parkinson = df[df.Pacientes == 'Parkinson']
df_control = df[df.Pacientes == 'Control']
df_ela = df[df.Pacientes == 'ELA']
df_huntington = df[df.Pacientes == 'Huntington']

#SD

#cont
print("CONTROL SD:")
_ , p_value_c = stats.shapiro(df_control["SD LSTI"])
print(f'LSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_control["SD RSTI"])
print(f'RSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_control["SD LSWI"])
print(f'LSWI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_control["SD RSWI"])
print(f'RSWI: {p_value_c}')
#park
print("PARKINSON SD:")
_ , p_value_c = stats.shapiro(df_parkinson["SD LSTI"])
print(f'LSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_parkinson["SD RSTI"])
print(f'RSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_parkinson["SD LSWI"])
print(f'LSWI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_parkinson["SD RSWI"])
print(f'RSWI: {p_value_c}')
#ela
print("ELA SD: ")
_ , p_value_c = stats.shapiro(df_ela["SD LSTI"])
print(f'LSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_ela["SD RSTI"])
print(f'RSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_ela["SD LSWI"])
print(f'LSWI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_ela["SD RSWI"])
print(f'RSWI: {p_value_c}')
#huntington
print("HUNTINGTON SD: ")
_ , p_value_c = stats.shapiro(df_huntington["SD LSTI"])
print(f'LSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_huntington["SD RSTI"])
print(f'RSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_huntington["SD LSWI"])
print(f'LSWI: {p_value_c}')
_ , p_value_c = stats.shapiro(df_huntington["SD RSWI"])
print(f'RSWI: {p_value_c}')
```

Como en los casos diferentes al control, el valor - p < 0.05, se rechaza H₀, por lo tanto, los datos no siguen una distribución normal. Por lo que se procede a hacer una transformada de potencia.

**1.1 Transformacion de potencia**

```python
print("CONTROL SD:")
_, lambd_LSTI = stats.boxcox(df_control["SD LSTI"])
_ , p_value_c = stats.shapiro(np.power(df_control["SD LSTI"],lambd_LSTI))
print(f'LSTI: {p_value_c}')
_, lambd_RSTI = stats.boxcox(df_control["SD RSTI"])
_ , p_value_c = stats.shapiro(np.power(df_control["SD LSWI"],lambd_RSTI))
print(f'RSTI: {p_value_c}')
_, lambd_LSWI = stats.boxcox(df_control["SD LSWI"])
_ , p_value_c = stats.shapiro(np.power(df_control["SD LSWI"],lambd_LSWI))
print(f'LSWI: {p_value_c}')
_, lambd_RSWI = stats.boxcox(df_control["SD RSWI"])
_ , p_value_c = stats.shapiro(np.power(df_control["SD RSWI"],lambd_RSWI))
print(f'RSWI: {p_value_c}')
#park
print("PARKINSON SD:")
_ , p_value_c = stats.shapiro(np.power(df_parkinson["SD LSTI"],lambd_LSTI))
print(f'LSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_parkinson["SD LSWI"],lambd_RSTI))
print(f'RSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_parkinson["SD LSWI"],lambd_LSWI))
print(f'LSWI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_parkinson["SD RSWI"],lambd_RSWI))
print(f'RSWI: {p_value_c}')
#ela
print("ELA SD: ")
_ , p_value_c = stats.shapiro(np.power(df_ela["SD LSTI"],lambd_LSTI))
print(f'LSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_ela["SD LSWI"],lambd_RSTI))
print(f'RSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_ela["SD LSWI"],lambd_LSWI))
print(f'LSWI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_ela["SD RSWI"],lambd_RSWI))
print(f'RSWI: {p_value_c}')
#huntington
print("HUNTINGTON SD: ")
_ , p_value_c = stats.shapiro(np.power(df_huntington["SD LSTI"],lambd_LSTI))
print(f'LSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_huntington["SD LSWI"],lambd_RSTI))
print(f'RSTI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_huntington["SD LSWI"],lambd_LSWI))
print(f'LSWI: {p_value_c}')
_ , p_value_c = stats.shapiro(np.power(df_huntington["SD RSWI"],lambd_RSWI))
print(f'RSWI: {p_value_c}')
```

Como se observa, la transformación de potencia no funcionó para la mayoría de métricas, por lo que se tendrá que hacer pruebas no paramétricas, sin embargo, se compara el control RSWI y parkinson RSWI las cuales siguieron una distribución normal después de la transformación.

**2. Prueba de independencia**

Son dos grupos *independientes* fisica y temporalmente, por que al ser distintos sujetos el grupo uno, dos, tres y cuatro, no dependen entre ellos

**3. Prueba de homocedasticidad**

Se le realiza la prueba, solo a aquellas que cumplieron con la distribucion normal luego de la transformacion

**Ho:** Las varianzas son iguales entre los grupos (homocedasticidad)

**H1:** Las varianzas no son iguales entre los grupos (heterocedasticidad)


```python
_, p_value = stats.levene(np.power(df_control["SD RSWI"],lambd_RSWI), np.power(df_parkinson["SD RSWI"],lambd_RSWI))
print(p_value)
```

Como el valor - p >  0.05, no se rechaza H₀, las varianzas son iguales (homocedasticidad)

Para este dato los tres supuestos se cumplen, se realiza la prueba t -test

**T-Test**

**Ho:** La media de las dos poblaciones es igual

**H1:** La media de las dos poblaciones no es igual


```python
_, p_value = stats.ttest_ind(np.power(df_control["SD RSWI"],lambd_RSWI), np.power(df_parkinson["SD RSWI"],lambd_RSWI))
print(p_value)
#Existe diferencia significativa
```

Como p- valor < 0.05 se rechaza Ho, lo que significa que las poblaciones tienen una diferencia estadisticamente significativa
